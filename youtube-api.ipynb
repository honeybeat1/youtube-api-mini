```python
import requests
```

```python
url = 'https://www.googleapis.com/youtube/v3/search'
params = dict(
    part = 'id,snippet',
    q = 'tom cruise', #query
    order = 'viewCount', #어느 순으로 정렬해서 보여줄 것인가 (relevance = 연관도)
    type = 'video', 
    maxResults = 50,
    key='AIzaSyBK7aaQgSA9mYbdQiJ5Eg6uu4oC_VdmD_s'
)
#불러오긔
response = requests.get(url=url, params=params)
data = response.json()
```

```python
def get_video_info(videoID):
    url1 = 'https://www.googleapis.com/youtube/v3/videos'
    
    params1 = dict(
        part = 'snippet, contentDetails, statistics', #가져올 내용 정하기
        id = videoID,
        key='AIzaSyBK7aaQgSA9mYbdQiJ5Eg6uu4oC_VdmD_s'
)

    response = requests.get(url=url1, params=params1)
    return response.json()

video = get_video_info(videoID)
video
```

```python
#영상 조회수 보기
video['items'][0]['statistics']['viewCount']
```

```python
#게시 날짜,시간
video['items'][0]['snippet']['publishedAt']
```

```python
html = []
for item in items:
    videoId = item['id']['videoId']
    videoInfo = get_video_info(videoId)
    viewCount = videoInfo['items'][0]['statistics']['viewCount']
    viewDate = videoInfo['items'][0]['snippet']['publishedAt']
    
    html.append('<li>') #리스트 태그 달아주기 (각각 아이템 하나당 하나의 리스트로 다루기 위해)
    html.append('<a href="https://www.youtube.com/watch?v={}">'.format(item['id']['videoId'])) #비디오에 바로가기 링크 달아주기
    html.append('<img src="{}">'.format(item['snippet']['thumbnails']['medium']['url']))#html태그에 img 태그 달아주기
    html.append('</a>') #위에 a태그 닫아주기 
    html.append('<h1>{}</h1>'.format(item['snippet']['title']))
    html.append('<span> 날짜 : {}</span>'.format(item['snippet']['publishedAt']))
    html.append('<span> 조회수 : {}</span>'.format(viewCount))
    html.append('</li>') #닫아주기
html
```

```python
temp = "\n".join(html) #리스트 형태면 뒤에서 깨짐
content = '<ul>{}</ul>'.format(temp)
content[:1000]
```
